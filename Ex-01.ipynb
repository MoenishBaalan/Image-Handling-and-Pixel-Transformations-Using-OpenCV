10. Read in the image ('Apollo-11-launch.jpg').

```python
apollo_img = cv2.imread('Apollo-11-launch.jpg')
apollo_img = cv2.cvtColor(apollo_img, cv2.COLOR_BGR2RGB)
```

11. Add the following text to the dark area at the bottom of the image (centered on the image):

```python
text = 'Apollo 11 Saturn V Launch, July 16, 1969'
font_face = cv2.FONT_HERSHEY_PLAIN
height, width, _ = apollo_img.shape
(text_w, text_h), baseline = cv2.getTextSize(text, font_face, 2, 2)
x = (width - text_w) // 2
y = height - 50 
cv2.putText(apollo_img, text, (x, y), font_face, 2, (255, 255, 255), 2, cv2.LINE_AA)
plt.imshow(apollo_img)
plt.axis('on')
plt.show()
```

12. Draw a magenta rectangle that encompasses the launch tower and the rocket.

```python
x1, y1 = 500, 50     
x2, y2 = 700, 600    
cv2.rectangle(apollo_img, (x1, y1), (x2, y2), (255, 0, 255), 3)
plt.imshow(apollo_img)
plt.axis('off')
plt.show()
```

13. Display the final annotated image.

```python
plt.imshow(apollo_img)
plt.axis("off")
plt.show()
```

14. Read the image ('boy.jpg').

```python
img_boy = cv2.imread("boy.jpg")
```

15. Adjust the brightness of the image.

```python
bright_boy = cv2.convertScaleAbs(img_boy, alpha=1, beta=50)
```

16. Create brighter and darker images.

```python
import numpy as np
matrix = np.ones(img_boy.shape, dtype="uint8") * 50  
img_brighter = cv2.add(img_boy, matrix)
img_darker = cv2.subtract(img_boy, matrix)
```

17. Display the images (Original Image, Darker Image, Brighter Image).

```python
plt.figure(figsize=(12,4))
plt.subplot(1,3,1)
plt.imshow(cv2.cvtColor(img_boy, cv2.COLOR_BGR2RGB))
plt.title("Original")
plt.axis("off")
plt.subplot(1,3,2)
plt.imshow(cv2.cvtColor(img_darker, cv2.COLOR_BGR2RGB))
plt.title("Darker")
plt.axis("off")
plt.subplot(1,3,3)
plt.imshow(cv2.cvtColor(img_brighter, cv2.COLOR_BGR2RGB))
plt.title("Brighter")
plt.axis("off")
plt.show()
```

18. Modify the image contrast.

```python
matrix1 = np.ones(img_boy.shape, dtype="uint8") * 1
matrix2 = np.ones(img_boy.shape, dtype="uint8") * 1
img_higher1 = cv2.convertScaleAbs(img_boy, alpha=1.1, beta=0)
img_higher2 = cv2.convertScaleAbs(img_boy, alpha=1.3, beta=0)
```

19. Display the images (Original, Lower Contrast, Higher Contrast).

```python
plt.figure(figsize=(12,4))
plt.subplot(1,3,1)
plt.imshow(cv2.cvtColor(img_boy, cv2.COLOR_BGR2RGB))
plt.title("Original")
plt.axis("off")
img_lower = cv2.convertScaleAbs(img_boy, alpha=0.9, beta=0)
plt.subplot(1,3,2)
plt.imshow(cv2.cvtColor(img_lower, cv2.COLOR_BGR2RGB))
plt.title("Lower Contrast")
plt.axis("off")
plt.subplot(1,3,3)
plt.imshow(cv2.cvtColor(img_higher2, cv2.COLOR_BGR2RGB))
plt.title("Higher Contrast")
plt.axis("off")

plt.show()
```

20. Split the image (boy.jpg) into the B,G,R components & Display the channels.

```python
B, G, R = cv2.split(img_boy)
plt.figure(figsize=(12,4))
plt.subplot(1,3,1)
plt.imshow(B, cmap="gray")
plt.title("Blue Channel")
plt.axis("off")
plt.subplot(1,3,2)
plt.imshow(G, cmap="gray")
plt.title("Green Channel")
plt.axis("off")
plt.subplot(1,3,3)
plt.imshow(R, cmap="gray")
plt.title("Red Channel")
plt.axis("off")
plt.show()
```

21. _merged the R, G, B , display along with orginal image 

```python
img_merged = cv2.merge([B, G, R])
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.imshow(cv2.cvtColor(img_boy, cv2.COLOR_BGR2RGB))
plt.title("Original Image")
plt.axis("off")
plt.subplot(1,2,2)
plt.imshow(cv2.cvtColor(img_merged, cv2.COLOR_BGR2RGB))
plt.title("Merged Image (B+G+R)")
plt.axis("off")
plt.show()
```

22. Split the image into the H, S, V components &  Display the channels.

```python
hsv_img = cv2.cvtColor(img_boy, cv2.COLOR_BGR2HSV)
H, S, V = cv2.split(hsv_img)
plt.figure(figsize=(12,4))
plt.subplot(1,3,1)
plt.imshow(H, cmap="gray")
plt.title("Hue Channel")
plt.axis("off")
plt.subplot(1,3,2)
plt.imshow(S, cmap="gray")
plt.title("Saturation Channel")
plt.axis("off")
plt.subplot(1,3,3)
plt.imshow(V, cmap="gray")
plt.title("Value Channel")
plt.axis("off")
plt.show()
```
